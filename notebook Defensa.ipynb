```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectFromModel
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('data/df_cleaned.csv')
```

# Filtro df por posisión > Defensa

```python
df=df[df['player_positions']=='defensa']
```

```python
#eliminar columna player_positions
df.drop(["player_positions"],axis=1,inplace=True)
```

```python
df.head()
```

# Limpieza gk

```python
df['gk'] =df['gk'].replace("+","")
```

```python
df['gk'] = df['gk'].str.replace(r'\+', '', regex=True)
```

```python
df.gk.fillna(0,inplace=True)
```

```python
df['gk']=df.gk.astype(int)
```

```python
df['gk'].unique()
```

# Limpieza columnas

```python
df.drop(['sofifa_id','short_name','long_name','dob','body_type','club_team_id','club_name','league_name',
 'league_level','club_jersey_number','club_joined','club_contract_valid_until','nationality_id','nationality_name',
 'nation_team_id','work_rate','real_face','player_traits'],axis=1,inplace=True)
```

```python
pd.set_option('display.max_columns', None)
print(df.columns.tolist())
```

# Categorizando preferred_foot

```python
pierna=[]
for i in df.preferred_foot:
    if i=="Left":
        pierna.append(0)
    else:
        pierna.append(1)
```

```python
df.drop("preferred_foot",axis=1,inplace=True)
df["preferred_foot"]=pierna
```

```python
df.select_dtypes("int64")
```

# Categorizando Progresion anual > target

```python
target=[]
#elegimos el % de target para que sea 0 y 1
for i in df.progresion_anual:
    if i>0.1:
        target.append(1)
    else:
        target.append(0)
df["target"]=target
```

```python
X=df.drop(["progresion_anual","target"],axis=1)
y=df.target
```

# RandomForestClassifier

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
#seleccion de columnas 
sel = SelectFromModel(RandomForestClassifier(n_estimators = 100))
sel.fit(X,y)
```

```python
sel.get_support()
```

```python
selected_feat= X_train.columns[(sel.get_support())]
len(selected_feat)
```

```python
X=df[['overall', 'potential', 'value_eur', 'age', 'height_cm', 'weight_kg',
       'pace', 'physic', 'attacking_crossing', 'attacking_finishing',
       'attacking_heading_accuracy', 'attacking_short_passing',
       'attacking_volleys', 'skill_dribbling', 'skill_curve',
       'skill_fk_accuracy', 'skill_long_passing', 'skill_ball_control',
       'movement_acceleration', 'movement_agility', 'movement_reactions',
       'movement_balance', 'power_shot_power', 'power_jumping',
       'power_stamina', 'power_strength', 'power_long_shots',
       'mentality_aggression', 'mentality_interceptions', 'mentality_vision',
       'mentality_penalties', 'mentality_composure',
       'defending_marking_awareness', 'defending_standing_tackle',
       'defending_sliding_tackle', 'ldm', 'cdm', 'rdm', 'lcb', 'rcb']]
```

```python
selected_feat
```

```python
model = RandomForestClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
y_pred=model.predict(X_test)
```

# Undersampling

```python
from imblearn.under_sampling import RandomUnderSampler
```

```python
rus = RandomUnderSampler(random_state=0)
X_res, y_res =rus.fit_resample(X, y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size=0.2, random_state=0)
```

```python
model=RandomForestClassifier()
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
y_pred=model.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
```

# Oversampling

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
ros = RandomOverSampler(random_state=0)
X_res, y_res = ros.fit_resample(X, y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size=0.2, random_state=0)
```

```python
model=RandomForestClassifier()
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
y_pred=model.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
```
